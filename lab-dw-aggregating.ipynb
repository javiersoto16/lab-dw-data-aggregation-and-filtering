1. Create a new DataFrame that only includes customers who:
   - have a **low total_claim_amount** (e.g., below $1,000),
   - have a response "Yes" to the last marketing campaign.

```python
import pandas as pd

url = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis.csv"
df = pd.read_csv(url)
```

```python
df.info()
df.head()
```

```python
df.columns = df.columns.str.lower().str.replace(' ', '_')
df['response'] = df['response'].str.strip().str.lower()
```

```python
df = df.drop(columns=['unnamed:_0']) 
```

```python
new_df = df[
    (df['total_claim_amount'] < 1000) &
    (df['response'] == 'yes')
].copy()
```

```python
new_df.shape
new_df.head()
```

2. Using the original Dataframe, analyze:
   - the average `monthly_premium` and/or customer lifetime value by `policy_type` and `gender` for customers who responded "Yes", and
   - compare these insights to `total_claim_amount` patterns, and discuss which segments appear most profitable or low-risk for the company.

```python
df['response'] = df['response'].str.strip().str.lower()  
df_yes = df[df['response'] == 'yes'].copy()
```

```python
pivot_metrics = pd.pivot_table(
    df_yes,
    values=['monthly_premium_auto', 'customer_lifetime_value'],
    index='policy_type',
    columns='gender',
    aggfunc='mean'
).round(2)

pivot_metrics
```

```python
pivot_claims = pd.pivot_table(
    df_yes,
    values='total_claim_amount',
    index='policy_type',
    columns='gender',
    aggfunc='mean'
).round(2)

pivot_claims
```

Los segmentos Corporate Auto (ambos géneros) y Special Auto (hombres) son los más rentables y de menor riesgo.

Personal Auto presenta mayor riesgo de reclamos, aunque con CLV y primas altas.

La segmentación por género y tipo de póliza permite personalizar campañas y optimizar la rentabilidad del portafolio de clientes.

4. Find the maximum, minimum, and median customer lifetime value by education level and gender. Write your conclusions.

```python
clv_stats = df.groupby(['education', 'gender'])['customer_lifetime_value'].agg(['max', 'min', 'median']).round(2)
clv_stats
```

Para los clientes con nivel educativo Bachelor, la mediana del CLV es similar entre mujeres (5,640.51) y hombres (5,548.03), aunque los valores máximos son mayores en mujeres, mostrando que algunas generan un valor excepcionalmente alto. En el grupo College, la mediana de los hombres (6,005.85) es ligeramente superior a la de las mujeres (5,623.61), mientras que los máximos son altos para ambos géneros.

En el caso de los clientes con nivel Doctor, la mediana se mantiene similar entre ambos géneros (F: 5,332.46, M: 5,577.67), aunque los valores máximos son menores que en otros grupos, indicando menos variabilidad en los extremos. Por otro lado, los clientes con educación High School or Below muestran medianas superiores (F: 6,039.55, M: 6,286.73) y un valor máximo muy alto en hombres (83,325.38), lo que demuestra que un nivel educativo menor no necesariamente implica un CLV bajo. Finalmente, los clientes con nivel Master presentan valores medianos y máximos equilibrados entre mujeres y hombres, con medianas de 5,729.86 y 5,579.10 respectivamente.

En conclusión, existe gran variabilidad dentro de cada grupo educativo, y no se observa una relación lineal entre mayor educación y CLV más alto. Las medianas son generalmente similares entre géneros, aunque los valores máximos pueden variar significativamente. Esto sugiere que la estrategia de fidelización y upselling debería centrarse en identificar a los clientes de alto CLV dentro de cada grupo, independientemente de su nivel educativo, y considerar otros factores como tipo de póliza, ingresos y canal de venta para una segmentación más efectiva.
